# STEP #3: CREATE TESTING AND TRAINING DATASET/DATA CLEANING

- Features are all nominal categorical variables so it is important to encode them with numbers to perform training with no specfific order.
- Values for veil color are brown, orange, white, and yellow, if you try to represent them as 1, 2, 3 and 4 would not make sense. Because this would imply that yellow is 4 times the value of brown.
- Note: for ordinal categorical variables, how many rooms in a house for example (3 rooms are more than 1) 

```python
# Let's test if there is any Null values!
total_null_values = sum(mushroom_df.isnull().sum())
print(total_null_values)
```

```python
# Let's drop the veil type (has only one class)
mushroom_df=mushroom_df.drop(["veil-type"],axis=1)
```

```python
mushroom_df.shape
```

```python
# from sklearn.preprocessing import LabelEncoder

# labelencoder = LabelEncoder()
# for column in mushroom_df.columns:
#     mushroom_df[column] = labelencoder.fit_transform(mushroom_df[column])
```

```python
mushroom_df
```

```python
y = mushroom_df['class']
X = mushroom_df.drop(['class'], axis=1)
X = pd.get_dummies(X, drop_first = True)
y = pd.get_dummies(y, drop_first = True)
```

```python
X
```

```python
y
```

```python
# from sklearn.preprocessing import StandardScaler

# scaler = StandardScaler()
# X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

# STEP#4: TRAINING THE MODEL

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train,y_train)
```

```python
type(X_train)
```

```python
feature_importances = pd.DataFrame(decision_tree.feature_importances_,
                                   index = X_train.columns,
                                   columns=['importance']).sort_values('importance',ascending=False)
```

```python
feature_importances
```

# STEP#5: EVALUATING THE MODEL 

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_predict_train = decision_tree.predict(X_train)
y_predict_train
cm = confusion_matrix(y_train, y_predict_train)
sns.heatmap(cm, annot=True)
```

```python
# Predicting the Test set results
y_predict_test = decision_tree.predict(X_test)
cm = confusion_matrix(y_test, y_predict_test)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict_test))
```

# STEP#6: IMPROVING THE MODEL 

```python
from sklearn.ensemble import RandomForestClassifier
RandomForest = RandomForestClassifier(n_estimators=150)
RandomForest.fit(X_train, y_train)
```

```python
y_predict_train = RandomForest.predict(X_train)
y_predict_train
cm = confusion_matrix(y_train, y_predict_train)
sns.heatmap(cm, annot=True)
```

```python
# Predicting the Test set results
y_predict_test = RandomForest.predict(X_test)
cm = confusion_matrix(y_test, y_predict_test)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict_test))
```
